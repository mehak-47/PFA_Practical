# Part A:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

column_names = ['ID', 'Math', 'Gender', 'StudyHours', 'Science']
df = pd.read_csv('mystery_data.csv', header=None, names=column_names)
df.head()
```

Justification 

ID cuz it has unique values

math because double numbers

gender because G and M 

Study Hours cuz in point value and single vals

Science cuz Double numbers

# Part B:

```python
df.dropna(subset = ['StudyHours'])`
```

```python
df['TotalScore'] = df['Math'] + df['Science']
df
```

```python
df.sort_values('TotalScore', ascending = False)
df.head(3)
```

# Part C

# 1

```python
# H0: = 70
# H1:!= 70
stud_hrs = df[df['StudyHours'] > 10]['Math']
stud_hrs
```

```python
ttest, p_value = stats.ttest_1samp(stud_hrs,70)

print(p_value)

if p_value < 0.05:
    print("We are rejecting Null hypothesis")
else:
    print("We are accepting Null hypothesis")
```

```python
sns.boxplot(x='Gender', y='TotalScore', data=df, color = 'yellow')
plt.xlabel('Gender')
plt.ylabel('Total Score')
plt.title('Score Distribution by Gender')
plt.show()
```
